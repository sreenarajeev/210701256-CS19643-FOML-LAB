```python
#remove column from dataset
df=df.drop(['Location'],axis=1)
df.info()
```

```python
from datetime import date
date.today().year
df['Age']=date.today().year-df['Previous Purchases']
df.head()
```

```python
df=df.drop(['Car_Age'],axis=1)
df.head()
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print(num_cols)
```

```python
for col in num_cols:
 print(col)
 print('Skew :', round(df[col].skew(), 2))
 plt.figure(figsize = (15, 4))
 plt.subplot(1, 2, 1)
 df[col].hist(grid=False)
 plt.ylabel('count')
 plt.subplot(1, 2, 2)
 sns.boxplot(x=df[col])
 plt.show()
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=df.drop(['Age','Previous Purchases'],axis=1))
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(df.drop(['Age','Previous Purchases'],axis=1).corr(), annot = True, vmin = -1, vmax = 1)
plt.show()
```
